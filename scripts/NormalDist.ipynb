```python
dist = rng.normal(0.0, 0.2, size=1000)
```

```python
fig, ax = plt.subplots(1,1, figsize=(5,7))
```

```python
x = plt.hist(dist, bins=20, density=True,label=r'$\mu=0.0 \sigma=0.2$')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
dist2 = rng.normal(0.0, 0.4, size=1000)
```
