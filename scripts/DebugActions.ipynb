```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('../logs/03-05/info9000.pickle', 'rb') as fp:
    data = pickle.load(fp)
```

```python
actiondata = data['actions']
```

```python
del data
```

```python
len(actiondata)
```

```python
actiondata[0]
```

```python
actions = np.array(actiondata)
```

```python
actions.shape
```

```python
del actiondata
```

```python
fig, axs = plt.subplots(2,2,figsize=(15,7))
axs[0,0].plot(actions[:,0], '^b', label='rotor 1')
axs[0,0].set_xlabel('Timestep')
axs[0,0].set_ylabel('Rotor Command')
axs[0,0].legend()
axs[0,1].plot(actions[:,1], 'vg', label='rotor 2')
axs[0,1].set_xlabel('Timestep')
axs[0,1].set_ylabel('Rotor Command')
axs[0,1].legend()
axs[1,0].plot(actions[:,2], '<r', label='rotor 3')
axs[1,0].set_xlabel('Timestep')
axs[1,0].set_ylabel('Rotor Command')
axs[1,0].legend()
axs[1,1].plot(actions[:,3], '>c', label='rotor 4')
axs[1,1].set_xlabel('Timestep')
axs[1,1].set_ylabel('Rotor Command')
axs[1,1].legend()
plt.show()
```

```python
plt.boxplot(actions, vert=False)
plt.ylabel('Rotors')
plt.xlabel('Control Value')
plt.show()
```
