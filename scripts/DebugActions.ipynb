```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
```

```python
with open('../logs/24-05/info20000.pickle', 'rb') as fp:
    data = pickle.load(fp)
```

```python
actiondata = data['actions']
```

```python
del data
```

```python
# fig, axs = plt.subplots(2,2,figsize=(15,7))
plt.plot(actions[:,0], '^b', label='rotor 1')
plt.xlabel('Timestep')
plt.ylabel('Rotor Command')
plt.legend()
plt.savefig('20k-1.png', format='png', dpi=300)
plt.show()
```
